#### 8. 사용자로 부터 정수를 입력받아서 1부터 그 사이에 존재하는 소수 (Prime number)를 출력하는 파이썬 프로그램을 작성하시오.

```python
int_input2 = input("정수를 입력하세요 : ")
```

```python
int_input2 = int(int_input2)
judge = 0

for i in range(2, int_input2):      ## 1~ 입력받은수 사이에서
    
    for j in range (2, i):          ##  소수인지 아닌지 판단하기
        
        if i % j == 0:          ## 나머지가 0이나오면 judge에 1을더하고 바로 break. judge가 1이 되면 소수가 아니고 끝까지 0이면 소수
            judge += 1
            break;
    
    if judge == 0 :
        print(i)
        
    judge = 0
```

#### 9. Kevin과 Stuart는 게임을 한다. 규칙은 아래와 같다.

* 두 플레이어에게 같은 string, S, 가 주어진다.
* 두 플레이어는 string S의 문자를 사용하여 substring을 만들어야 한다(중복 가능).
* Stuart는 자음으로 시작하는 단어를 만들어야 한다.
* Kevin은 모음으로 시작하는 단어를 만들어야 한다.
* 게임은 두 플레이어가 만들 수 있는 모든 단어를 만들었을 때 끝난다.
* 점수 계산 : substring 1개당 +1 점을 획득한다. (아래 그림 참고)

```python
S = input ("string S를 입력하세요(대문자로 입력!) : ")
```

```python
S_len = len(S)
score_Stuart = 0
score_Kevin = 0

for i in range(S_len) :
    if S[i] == 'A' or S[i] == 'E' or S[i] == 'I' or S[i] == 'O' or S[i] == 'U' : ## 모음이면
        score_Kevin += (S_len - i)
        
    else : ## 모음이 아니면 -> 즉 자음이면
        score_Stuart += (S_len - i)
```

```python
score_Stuart
```

```python
score_Kevin
```

```python
if score_Stuart > score_Kevin :
    print ("승자 Stuart 점수 :", score_Stuart)
elif score_Stuart < score_Kevin :
    print ("승자 Kevin 점수 :", score_Kevin)
else :
    print ("무승부 , 점수 :", score_Stuart)
```

#### 간략한 설명
    문자열에서 한글자 한글자씩 따온다는 느낌으로 해야한다. 첫번째 글자 B로 만들 수 있는건 BA, BAN, BANA, BANAN, BANANA
    두번째 글자 A는 모음이고, 그다음 자음 N으로 만들 수 있는건 N, NA, NAN, NANA가 된다. 
    마지막 자음 N으로 만들 수 있는건 N하고 NA가 된다. 
    
    이렇게 나오는 각각의 CASE들은 [문자열의 길이 - 인덱스값]에 해당하므로, 이 CASE들의 합을 구해주면 된다.

#### 10. 알파벳 소문자만으로 이루어진 string, S, 가 주어지고 아래 2가지 연산을 수행할 수 있다.
* 추가 : string S의 끝에 알파벳 소문자 하나를 추가한다.
* 삭제 : string S의 마지막 문자 하나를 삭제한다. empty string에서 삭제 연산을 수행한 결과는 empty string이다.

두 개의 string, s와 t, 그리고 정수 k가 주어질 때, 정확히 k번의 연산을 통해 s를 t로 바꿀 수 있는지 판단하는 함수를 작성하시오.



```python
def appendDelete(s, t, k):
    s_len = len(s)
    t_len = len(t)
    min_k = 0             ## k가 될수있는 최솟값 , 이게 있어야지 Yes or no를 판단할 수 있다.
    count = 0              ## 글자가 몇번째 인덱스 까지 겹쳐있는지 카운트하는 변수
    
    for i in range(s_len):      ## count값을 구하는 반복문
        if s[i] == t[i]:
            count += 1
        else :
            break
        
    min_k = (s_len - count) + (t_len - count)
    
    if k>= s_len + t_len :                 ## empty칸까지 갔다가 추가연산하는 경우
        return 'Yes'
    elif k>= min_k and k%2 == min_k%2 :     ## 나머지 부분에서, k가 min_k 이상이고, k+2 일경우를 이렇게 표현해보았다.
        return 'Yes'
    else:
        return 'No'
    
    
    return 'Yes'    # or 'No'
```

```python
s = 'hackerhappy'
t = 'hackerrank'

appendDelete(s, t, 9)
```

```python
appendDelete(s, t, 10)
```

```python
appendDelete(s, t, 24)
```

```python
appendDelete(s, t, 17)
```

#### 간략한 설명

    k가 될 수있는 최솟값(min_k)을 구해야 한다. 그래야지 Yes or no를 판단할 수 있다.
    min_k는 (겹쳐있는곳 까지 삭제할 때 수행되는 연산) + (다시 t까지 추가할 때 수행되는 연산) 이다.
    min_k +2 , min_k +4 , min_k +6 같은 경우에는 당연히 'Yes'이다. 왜냐면 삭제했다가 하나 추가하면 되기때문에
    그런데 s를 전부삭제하고 empty칸에서 삭제연산을 하는경우를 고려해야한다. 그 때부터는 어느 숫자든 될 수 있다.
    참고로 empty칸을 만들려면 len(s)만큼의 삭제연산이 필요하다. 그리고 다시채울려면 len(t)만큼의 추가연산이 필요하다.
    즉 k가 len(s) + len(t) 이상이 되면 어떤숫자가 되든 가능하다는 얘기이다.
