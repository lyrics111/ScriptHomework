### 7. 다음 문제를 정확히 읽고 적합한 코딩을 제시하시오
* length(X)를 다음과 같이 정의한다 : 10진수로 표현된 X의 자릿수
* strange라는 숫자를 아래과 같이 정의한다.    
    * 음이 아닌 정수 중 자릿수가 1인 모든 숫자는 Strange이다.
    * length(X) > 1 인 숫자 X 는 다음 2가지를 만족하면 Strange이다.
        * X는 length(X)로 정확히 나누어 떨어진다.
        * X / length(X) 가 strnage이다.

* 이 문제에서 숫자 하나가 주어질 때, 이 숫자가 Strange인지 판단하는 함수 isStrange(num)을 작성하시오.
* isStrange(num) 함수가 재귀적인 함수가 되도록 작성해 보세요.

```python
def isStrange(num):
    length = 0
    temp = num # 임시로 num 을 저장할 변수
    
    if num == 0: #0이면 바로 True 리턴 , while문에서 에러방지
        return True 
    
    while (num != 0) : #length값 구하기
        num = num//10 
        length +=1

    if length == 1 :   #length가 1인 모든숫자는 strange
        return True
    
    num = temp # while문에서 값이 변한 num을 다시 원상복귀
    
    if num % length != 0 : #num이 length로 나누어 떨어지지 않으면 False
        return False
    else :  #num이 length로 나누어 떨어지면 다시 isStrange함수를 호출하여 ( num // length ) 를 대입
        return isStrange(num // length)
        
```

```python
print(isStrange(0))
print(isStrange(56))
print(isStrange(100))
print(isStrange(7882602506492313600))
print(isStrange(601230254952209865))
```

### 8. 다음 문제를 정확히 읽고 적합한 코딩을 제시하시오
* 1번 그룹, 2번 그룹, ... n번 그룹으로 번호가 매겨진 총 n개의 그룹이 존재하고, ith 그룹에는 임의의 팀원 수 ai 명이 있다.
  하나의 버스 정류장이 있고, 버스는 오직 한 대만 운영한다. 버스는 버스 정류장에 도착하고 모든 그룹은 버스를 통해 여행하길 원한다.
  하지만, 같은 그룹에 속한 팀원들은 분리되기를 원치 않는다. 그래서 그룹의 모든 팀원이 탈 수 있을 때 에만 그 그룹은 버스에 탑승한다.
  또한, 이들은 그룹의 번호순으로 여행을 하길 원한다. 예를 들어, 3번 그룹은 1번 그룹과 2번 그룹이 여행을 한 뒤에 버스를 탈 수 있거나, 1   번 그룹과 2번 그룹이 이미 버스에 탑승한 상태에 있을 때 동시에 버스를 탈 수 있다.
  마지막으로, 버스는 빈 좌석이 없을 때 에만 여행을 떠난다.

* 이 문제에서 각 그룹의 팀원 수가 리스트로 주어질 때, 가능한 모든 버스의 사이즈 x(x명의 사람들이 동시에 탈 수 있음)를 반환하는 함수     solve(a) 를 작성하시오.
* [주의] 반드시 계산 복잡도를 최대한 낮추세요. 아래 예시 3번을 실행했을 때 결과가 바로 나타날 수 있도록!!

```python
def solve(a):
    result = []
    #max = sum(a)  #가장 좌석이 많은 사이즈의 버스. 모든승객들을 한대에 몽땅 태우고가는 경우의수. 무조건 존재하는 경우의수.
    #print(max)
    #print()
    
    min = a[0]    #버스사이즈는 리스트내의 가장 큰 수보다는 크거나 같아야한다. 따라서 for문으로 리스트 내의 가장 큰수를 구한다
    for i in range(1, len(a)) :
        if a[i] > min :
            min = a[i]  # min은 리스트내의 가장 큰수이자 이 버스의 최소사이즈. 
            
    for i in range(len(a)):
        case = sum(a[0:i+1])     # 앞에서부타 순차적으로 더해갔을때
        if case >= min :      # case가 min 이상의 값이되면은 경우의수로 의심을 해본다.
            #print(case)
            if sum(a) % case == 0 :    #리스트의 총합을 case로 나눠서 나누어떨어지면은 알맞은 경우의 수이다.
                result.append(case)    #그 경우의 수를 결과리스트 result에 추가
    return result
```

```python
a = [1,3,5,2,1,4,7,2,21,23,17,6]
solve(a)
```

```python
## 순수문자열을 도출하는 함수 (공백이랑 구두문자 포함)
def get_pure_text (str) :
    HTML_flag = False
    style_flag = False
    script_flag = False

    judge = [HTML_flag , style_flag, script_flag]
    result = []  #문자열들을 리스트에 넣은후에 join을 이용해서 다시 문자열로 반환
    for i in range (len(str)):

        if str[i] == '<' :
            judge[0] = True
                
         # i=0일때 i-1이 안될거라 생각했는데 생각해보니 파이썬에서 str[-1]은 가장 뒤의 인덱스다. 따라서 에러가 뜨지않는다.
                
        if str[i-1] == '>' and str[i] != '<':
            judge[0] = False

        if str[i:i+6] == '<style' :
            judge[1] = True

        if str[i:i+8] == '</style>' :
            judge[1] = False
        
        if str[i:i+7] == '<script' :
            judge[2] = True
        
        if str[i:i+9] == '</script>' :
            judge[2] = False

        if True in judge :    ## 만약에 하나의 Flag라도 True를 나타낸다면 문자열을 대입하지않는다
            continue
            
        result.append(str[i])
    return ''.join(result)

## 공백을 기준으로 문자열의 단어를 세는 함수 생성
def get_word_count (str) : #문자열에서 구두문자, 공백문자 제거하고 사전형태로 출력하는 함수
    s = ''.join([i for i in str if i not in string.punctuation])  # 구두문자가 제거된 문자열
    list = []
    list = s.split() #공백제거된 문자열을 리스트로 반환
    word_dict = {} 
    
    for word in list :
        if word not in word_dict :  #word_dict에 list에 원소가 없으면 value에 1을지정하고
            word_dict[word] = 1
        else :
            word_dict[word] += 1    #이미 단어가 들어와있으면 1을더한다
    return word_dict
```
